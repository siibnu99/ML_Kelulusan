```python
def fetch_data_with_columns(table_name, limit=10, offset=0):

    # Eksekusi query untuk mendapatkan semua data dan nama kolom dari tabel
    query = f"SELECT * FROM {table_name} LIMIT {limit} OFFSET {offset}"
    cursor.execute(query)

    # Ambil nama kolom dari cursor.description
    column_names = [desc[0] for desc in cursor.description]

    # Ambil semua data
    data = cursor.fetchall()

    # Tutup koneksi
    cursor.close()

    return column_names, data
```

```python
# remove kode_tahun_ajaran except 20191, 20192, 20201, 20202, 20211, 20212, 20221, 20222
df = df[df['kode_tahun_ajaran'].isin(['20191', '20192', '20201', '20202', '20211', '20212', '20221', '20222', '20231'])]
```

```python
# foreach data all
for column in df.columns:
    print(f"Column {column} has {df[column].nunique()} unique values")
```

```python
data = {}
for index, row in df.iterrows():
    result = data.get(row['id_mahasiswa'])
    if not (result is not None and result.any()):
        data[row['id_mahasiswa']] = row
    row['ips'] = round(row['ips'], 2)
    data[row['id_mahasiswa']]['id_status_mahasiswa'] = 'L'
    if row['kode_tahun_ajaran'] == '20191':
        data[row['id_mahasiswa']]['IPS1'] = row['ips']
        data[row['id_mahasiswa']]['SKS1'] = row['sks']
    elif row['kode_tahun_ajaran'] == '20192':
        data[row['id_mahasiswa']]['IPS2'] = row['ips']
        data[row['id_mahasiswa']]['SKS2'] = row['sks']
    elif row['kode_tahun_ajaran'] == '20201':
        data[row['id_mahasiswa']]['IPS3'] = row['ips']
        data[row['id_mahasiswa']]['SKS3'] = row['sks']
    elif row['kode_tahun_ajaran'] == '20202':
        data[row['id_mahasiswa']]['IPS4'] = row['ips']
        data[row['id_mahasiswa']]['SKS4'] = row['sks']
    elif row['kode_tahun_ajaran'] == '20211':
        data[row['id_mahasiswa']]['IPS5'] = row['ips']
        data[row['id_mahasiswa']]['SKS5'] = row['sks']
    elif row['kode_tahun_ajaran'] == '20212':
        data[row['id_mahasiswa']]['IPS6'] = row['ips']
        data[row['id_mahasiswa']]['SKS6'] = row['sks']
    elif row['kode_tahun_ajaran'] == '20221':
        data[row['id_mahasiswa']]['IPS7'] = row['ips']
        data[row['id_mahasiswa']]['SKS7'] = row['sks']
    elif row['kode_tahun_ajaran'] == '20222':
        data[row['id_mahasiswa']]['IPS8'] = row['ips']
        data[row['id_mahasiswa']]['SKS8'] = row['sks']
    elif row['kode_tahun_ajaran'] == '20231':
        data[row['id_mahasiswa']]['id_status_mahasiswa'] = 'N'
    else:
        print('unknown')

```
